### 1.1 Data Types & Control Flow

Write a function `classify_number(n)` that:
- Returns `"negative"` if `n < 0`
- Returns `"zero"` if `n == 0`
- Returns `"small positive"` if `0 < n <= 10`
- Returns `"large positive"` otherwise

````python
def classify_number(n):
    # I am checking the number step by step so the logic stays easy to read.
    if n < 0:
        return 'negative'
    elif n == 0:
        return 'zero'
    elif n <= 10:
        return 'small positive'
    else:
        return 'large positive'

print('1.1 Example values:')
print('classify_number(-5) ->', classify_number(-5))
print('classify_number(0)  ->', classify_number(0))
print('classify_number(7)  ->', classify_number(7))
print('classify_number(42) ->', classify_number(42))
````

````python
words = ['apple', 'banana', 'cherry', 'apple', 'date', 'banana', 'apple']

# I am building the frequency table manually first so the counting process is visible.
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] = word_count[word] + 1
    else:
        word_count[word] = 1

# A set is a simple way to keep only the unique items.
unique_words = set(words)

# I am using a loop here instead of a compact comprehension so the filter is obvious.
long_words = []
for word in words:
    if len(word) > 5:
        long_words.append(word)

print('1.2 Word count:')
print(word_count)
print('Unique words:')
print(unique_words)
print('Words longer than 5 characters:')
print(long_words)
````

### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    # I check the input types first so the function stays predictable for a student reading it.
    numeric_types = (int, float, np.number)
    if not isinstance(a, numeric_types) or not isinstance(b, numeric_types):
        raise TypeError('Inputs must be numeric')

    if b == 0:
        return None

    return a / b

print('1.3 Example values:')
print('safe_divide(10, 2) ->', safe_divide(10, 2))
print('safe_divide(5, 0)  ->', safe_divide(5, 0))
print('If a non-numeric input is used, the function raises TypeError.')
````

````python
def apply_twice(f, x):
    # The idea is simple: send the value through the same function two times.
    first_pass = f(x)
    second_pass = f(first_pass)
    return second_pass

# I am using a normal function here so the notebook reads more like notes.
def triple(x):
    return x * 3

result = apply_twice(triple, 4)
print('1.4 Example result:')
print('apply_twice(triple, 4) ->', result)
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

# I use eig() to see how this matrix acts along its special directions.
eigenvalues, eigenvectors = np.linalg.eig(A)

print('4.3 Eigenvalues:')
print(eigenvalues)
print('4.3 Eigenvectors stored as columns:')
print(eigenvectors)

print('4.3 Checking A @ v against scaling by the eigenvalue for each eigenpair:')
for i in range(len(eigenvalues)):
    vector = eigenvectors[:, i]
    left_side = A @ vector
    right_side = eigenvalues[i] * vector
    difference = np.linalg.norm(left_side - right_side)
    print('Eigenpair', i + 1, 'difference:', difference)

fig, ax = plt.subplots()
colors = ['tab:blue', 'tab:orange']
for i in range(len(eigenvalues)):
    vector = eigenvectors[:, i]
    scaled_vector = eigenvalues[i] * vector
    ax.arrow(0, 0, scaled_vector[0], scaled_vector[1], head_width=0.15, head_length=0.15, length_includes_head=True, color=colors[i])
    ax.text(scaled_vector[0], scaled_vector[1], f'pair {i + 1}')

ax.set_xlim(-5, 6)
ax.set_ylim(-5, 6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Eigenvectors scaled by eigenvalues')
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)
plt.tight_layout()
plt.show()
````

**Your Geometric Explanation (4.3):**

> When a matrix acts on a vector it can change its direction and length. An eigenvector is a special direction that the matrix only stretches or shrinks (by the eigenvalue) without changing direction. Geometrically, eigenvalues describe how much scaling occurs along those invariant directions.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

# I split SVD into pieces so I can rebuild the matrix and also make a simple rank-1 version.
U, S, Vt = np.linalg.svd(X, full_matrices=False)

sigma = np.diag(S)
X_reconstructed = U @ sigma @ Vt

u1 = U[:, 0].reshape(-1, 1)
v1 = Vt[0, :].reshape(1, -1)
X_approx = S[0] * (u1 @ v1)

print('4.4 Shape of U:', U.shape)
print('4.4 Shape of S:', S.shape)
print('4.4 Shape of Vt:', Vt.shape)
print('4.4 Singular values:')
print(S)
print('4.4 Reconstruction error:')
print(np.linalg.norm(X - X_reconstructed))
print('4.4 Rank-1 approximation:')
print(X_approx)
````

**SVD → PCA Connection (4.4):**

> `U` contains left singular vectors, `S` contains singular values (importance), and `Vt` rows are principal directions. PCA on centered data yields principal components in `Vt` (rows) and singular values relate to explained variance. The directions of maximum variance are in `Vt`.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# I compare both samples against a normal distribution to see how different they are.
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print('5.4 KS test for s1 (normal sample):')
print('statistic:', ks_s1.statistic)
print('p-value:', ks_s1.pvalue)
print('5.4 KS test for s2 (exponential sample):')
print('statistic:', ks_s2.statistic)
print('p-value:', ks_s2.pvalue)

# I build a simple trending series so the ADF test has something non-stationary to detect.
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)
adf_result = adfuller(ts)

print('5.4 ADF statistic for original series:', adf_result[0])
print('5.4 ADF p-value for original series:', adf_result[1])

# Differencing removes the trend, so I test the new series again.
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)

print('5.4 ADF statistic after differencing:', adf_diff[0])
print('5.4 ADF p-value after differencing:', adf_diff[1])
if adf_diff[1] < 0.05:
    print('5.4 The differenced series looks stationary.')
else:
    print('5.4 The differenced series still does not look stationary.')
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    eps = 1e-10
    bins_edges = np.linspace(min(expected.min(), actual.min()), max(expected.max(), actual.max()), bins + 1)
    exp_counts, _ = np.histogram(expected, bins=bins_edges)
    act_counts, _ = np.histogram(actual, bins=bins_edges)
    exp_pct = exp_counts / exp_counts.sum()
    act_pct = act_counts / act_counts.sum()
    # avoid zeros
    exp_pct = exp_pct + eps
    act_pct = act_pct + eps
    psi = np.sum((act_pct - exp_pct) * np.log(act_pct / exp_pct))
    return psi

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
ax.hist(train_dist, bins=30, density=True, alpha=0.6, label='train')
ax.hist(drift_dist, bins=30, density=True, alpha=0.6, label='prod')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** Change in the relationship between inputs and target.
- **Covariate drift:** Change in the input feature distribution.
- **PSI < 0.1:** stable/no meaningful shift.
- **PSI 0.1–0.2:** minor shift.
- **PSI > 0.2:** major shift.
- **Retraining trigger example:** sustained PSI > 0.2 or drop in performance metric over time.

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red = 4
blue = 3
green = 3

# I start with the basic marble probabilities before moving to the two-draw case.
p_red = red / total
p_blue = blue / total
p_green = green / total

p_red_then_blue = (red / total) * (blue / (total - 1))
p_blue_given_red = blue / (total - 1)
independent = p_blue_given_red == p_blue

print('6.1 P(red):', p_red)
print('6.1 P(blue):', p_blue)
print('6.1 P(green):', p_green)
print('6.1 P(red then blue):', p_red_then_blue)
print('6.1 P(blue given red):', p_blue_given_red)
print('6.1 Are the draws independent?', independent)
if independent:
    print('6.1 The draws would be independent.')
else:
    print('6.1 The draws are not independent because the first draw changes the second probability.')
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# I use the law of total probability first, then plug that result into Bayes' theorem.
p_free = p_free_given_spam * p_spam + p_free_given_ham * p_ham
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    p_ham = 1 - prior_spam
    p_word = p_word_given_spam * prior_spam + p_word_given_ham * p_ham
    return (p_word_given_spam * prior_spam) / p_word

pred = naive_bayes_predict(0.30, 0.80, 0.05)
print('6.3 naive_bayes_predict result:', pred)
print('6.3 Difference between manual and function result:', abs(pred - p_spam_given_free))
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | 0.30 |
| Likelihood | P(word \| Spam) | 0.80 |
| Evidence | P(word) | computed p_free |
| Posterior | P(Spam \| word) | computed p_spam_given_free |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

# I repeatedly sample from the population so I can see how the sample mean behaves.
n_samples = 5000
sample_size = 30
sample_means = []
for _ in range(n_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(sample.mean())

sample_means = np.array(sample_means)

pop_mean = population.mean()
pop_std = population.std()
clt_std = pop_std / np.sqrt(sample_size)

print('6.4 Population mean:', pop_mean)
print('6.4 Population std:', pop_std)
print('6.4 Sample means mean:', sample_means.mean())
print('6.4 Sample means std:', sample_means.std())
print('6.4 CLT predicted std:', clt_std)

fig, ax = plt.subplots()
ax.hist(sample_means, bins=30, density=True, alpha=0.7, label='Sample means')
xs = np.linspace(sample_means.min(), sample_means.max(), 200)
ax.plot(xs, stats.norm.pdf(xs, pop_mean, clt_std), color='tab:orange', label='Theoretical normal')
ax.set_title('CLT: Distribution of Sample Means')
ax.set_xlabel('Sample mean')
ax.set_ylabel('Density')
ax.legend()
plt.tight_layout()
plt.show()

ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print('6.4 KS test statistic:', ks_result.statistic)
print('6.4 KS test p-value:', ks_result.pvalue)
if ks_result.pvalue > 0.05:
    print('6.4 The sample means look approximately normal.')
else:
    print('6.4 The sample means do not look very normal.')
````

**CLT Reflection (6.4):**

> The CLT explains why sample means tend to be normally distributed regardless of population shape, which justifies using normal-based inference for many statistics when sample sizes are reasonable.
